**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

**Ans1:**

- **Matplotlib** is a popular 2D plotting library for the Python programming language.

- It provides a wide variety of static, animated, and interactive plots to visualize data.

- Matplotlib is often used for creating high-quality plots and charts for data analysis, exploration, and presentation.

- The **Pyplot** module is a collection of functions in Matplotlib that make it easy to create plots with fewer lines of code. It provides a MATLAB-like interface for creating plots and is widely used in scientific computing, data analysis, and machine learning.

- Here are five types of plots that can be created using the Pyplot module of Matplotlib:
    
    **1.Scatter Plot:** 
    - *Function*: **plt.scatter()**
    - *Description*: A scatter plot displays individual data points on a two-dimensional graph, with each point representing values for two variables. It is useful for identifying relationships, clusters, or outliers in the data.

    **2.Line Plot:** 
    - *Function*: **plt.plot()**
    - *Description*: A line plot is used to visualize the relationship between two continuous variables by connecting data points with straight lines. It is commonly used to show trends or patterns over a range of values.
    
    **3.Bar Plot:**
    - *Function*: **plt.bar()**
    - *Description*: A bar plot represents categorical data with rectangular bars, where the length of each bar corresponds to the value it represents. It is effective for comparing and displaying values across different categories.
    
    **4.Histogram:** 
    - *Function*: **plt.hist()**
    - *Description*: A histogram is used to represent the distribution of a single variable by dividing the data into bins and displaying the frequency or probability of values falling into each bin. It provides insights into the data's underlying distribution.
    
    **5.Pie Chart:** 
    - *Function*: **plt.pie()**
    - *Description*: A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice represents a proportion of the whole, making it useful for displaying the composition of a categorical variable as parts of a whole.

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

**Note: Also add title, xlabel, and ylabel to the plot.**

**Ans2:**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
x
```

```python
y
```

```python
# Creating a scatter plot
plt.scatter(x,y,color='red',marker='.')  # '.' for point marker

plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show() # Show the plot
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:**

import numpy as np

**For line 1:** x = np.array([0, 1, 2, 3, 4, 5]) and  y = np.array([0, 100, 200, 300, 400, 500])

**For line 2:** x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

**For line 3:** x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

**For line 4:** x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

**Ans3:**

- The **subplot()** function in Matplotlib is used to create a grid of subplots within a single figure.
- It allows us to arrange multiple plots in rows and columns, making it convenient for visualizing and comparing different sets of data.
- The function takes three arguments: the number of rows, the number of columns, and the index of the subplot to create.

```python
# Data for four lines
# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Creating subplots 
plt.subplot(2, 2, 1)  # creates a 2x2 grid and selects the first subplot for plotting x1,y1
plt.plot(x1, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)  # selects the second subplot for plotting x2,y2
plt.plot(x2, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)  # selects the third subplot for plotting x3,y3
plt.plot(x3, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)  # selects the fourth subplot for plotting x4,y4
plt.plot(x4, y4)
plt.title('Line 4')

plt.show() # Show the plot
```

**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

**Ans4:**

- A **bar plot** (or **bar chart**) is a type of chart that presents categorical data with rectangular bars. Each bar's length or height corresponds to the value it represents. Bar plots are useful for comparing and displaying the values of different categories or groups.
- A **horizontal bar plot** is simply a bar plot with the bars displayed horizontally instead of vertically.

```python
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])

# Bar Plot
plt.figure(figsize=(8, 4))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Profit of Tech Companies (Bar Graph)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 4))
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Profit of Tech Companies (Horizontal Bar Graph)')
plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

**Ans5:**

- A **box plot**, also known as a **box-and-whisker plot**, is a graphical representation of the distribution of a dataset.
- It displays the summary statistics of a dataset, including the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.
- Box plots are useful for visualizing the spread, central tendency, and skewness of data.
- They are particularly effective for comparing the distribution of multiple datasets.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
box1
```

```python
box2
```

- **plt.boxplot()** is the function to draw a box and whisker plot. The box extends from the first quartile (Q1) to the third quartile (Q3) of the data, with a line at the median. The whiskers extend from the box to the farthest data point lying within 1.5x the inter-quartile range (IQR) from the box. Flier points are those past the end of the whiskers.

```python
# Creating a box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
