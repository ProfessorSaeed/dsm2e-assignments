**Ques 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**Ans 1:**

- Seaborn provides a variety of plots for different types of data analysis. Here are five commonly used plots along with their typical use cases:

    **1.Scatter Plot:**
    - **Use:** Shows the relationship between two variables by plotting points on a two-dimensional plane. Each point represents an observation.
    - **Seaborn Function:**
        - *__sns.scatterplot()__*: should be used when both variables are numeric, or 
        - *__sns.relplot()__*: scatterplot() is the default kind in relplot() (it can also be forced by setting kind="scatter")
    
    **2.Line Plot:**
    - **Use:** Displays the relationship between two numeric variables over a continuous interval. It's often used to visualize trends over time.
    - **Seaborn Function:**
        - *__sns.lineplot()__* or
        - *__sns.relplot()__* by setting kind="line"
    
    **3.Bar Plot:**
    - **Use:** Represents categorical data with rectangular bars. Useful for comparing the quantities of different categories.
    - **Seaborn Function:**
        - *__sns.barplot()__* or
        - *__sns.catplot()__* by setting kind="bar"
    
    **4.Box Plot (Box-and-Whisker Plot):**
    - **Use:** Displays the distribution of a dataset and provides a summary of key statistics such as median, quartiles, and outliers. Useful for detecting variations and comparing multiple distributions.
    - **Seaborn Function:**
        - *__sns.boxplot()__* or
        - *__sns.catplot()__* by setting kind="box"
    
    **5.Heatmap:**
    - **Use:** Represents data in a matrix format where values are represented by colors. Useful for visualizing the correlation between variables or displaying a grid of values.
    - **Seaborn Function:** *__sns.heatmap()__*

**Ques 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.**

**Ans 2:**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
fmri = sns.load_dataset('fmri')
fmri
```

```python
# Line plot using sns.relplot()
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line", hue="region", style="event")
plt.show()
```

```python
# Line plot using sns.lineplot()
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event")
plt.show()
```

**Ques 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.**

**Ans 3:**

```python
# Loading the dataset
titanic = sns.load_dataset('titanic')
titanic
```

```python
# Box Plot using sns.catplot()
# Box plot for 'age'
sns.catplot(data=titanic, x="pclass", y="age", kind="box")
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
sns.catplot(data=titanic, x="pclass", y="fare", kind="box")
plt.title('Box Plot of Fare by Pclass')

# Show the plots
plt.show()
```

```python
# Box Plot using sns.boxplot()

plt.figure(figsize=(12, 6))

# Box plot for 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')

# Box plot for 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')

plt.tight_layout() # Adjust layout

plt.show() # Show the plots
```

**Ques 6: Use the "flights" dataset from seaborn to plot a heatmap.**

**Ans 6:**

- The `sns.heatmap()` function in Seaborn is used to create a **heatmap**, which is a graphical representation of data in a matrix format where values are represented by colors. The function is versatile and allows us to visualize the relationships and patterns in the data. Here's an explanation of some important parameters of `sns.heatmap` function:

    **1.data:** The input data for the heatmap. This can be a 2D array, a DataFrame, or a pivot table.
    
    **2.cmap:** The color map to be used for mapping the values to colors. We can choose from a variety of predefined colormaps, such as 'viridis', 'YlGnBu', 'coolwarm', etc.
    
    **3.annot:** If True, the function will annotate each cell with the numeric value. This is useful for displaying the actual values in the cells.
    
    **4.fmt:** String formatting code to use when annotating cells. For example, 'd' for integers, '.2f' for floating-point numbers with two decimal places, etc.
    
    **5.linewidths, linecolor:** These parameters control the width and color of the lines that divide each cell.
    
    **6.cbar:** If True, a colorbar will be displayed alongside the heatmap, indicating the mapping of values to colors.
    
    **7.xticklabels, yticklabels:** If set to 'auto', Seaborn will try to infer whether to show tick labels based on the data. You can also pass a list of labels or False to hide the labels.
    
    **8.ax:** Specifies the axis on which the heatmap will be plotted. If not provided, the current active axis will be used.

```python
# Loading the dataset
flights = sns.load_dataset('flights')
flights
```

```python
# Pivot the dataset to create a matrix for heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)

plt.title('Heatmap: Flight Passenger Counts by Month and Year')
plt.show()
```
